```python
small_graph = {
    "S": ["A", "B"],
    "A": ["C", "D"],
    "B": ["G", "H"],
    "C": ["E", "F"],
    "D": [],
    "G": ["I"],
    "H": [],
    "E": ["K"],
    "F": [],
    "I": [],
    "K": []   
}

from collections import deque
def bfs(graph, start, goal):
    visited = set()
    queue = deque([start])
    while queue:
        node = queue.pop()
        print("on node "+ node)
        if node not in visited:
            visited.add(node)
            print("Node " +node+ " visited for first time")
            if node == goal:
                print("Goal: "+ node)
                return True
            for neighbor in graph[node]:
                print(neighbor +" is neighbor to "+ node)
                if neighbor not in visited:
                    print("Marking "+ neighbor +" neighbor to "+ node +" as not visited and add to left of its branch")
                    queue.appendleft(neighbor)
    return False
bfs(small_graph, 'S', 'H')
```
